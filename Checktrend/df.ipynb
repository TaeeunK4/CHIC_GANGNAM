```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('ebay_luxury_bags_sold_20260211.csv')
df_1 = pd.read_csv('pytrends_brands_timeline.csv')
df_2 = pd.read_csv('pytrends_colors_timeline.csv')
df_3 = pd.read_csv('pytrends_types_timeline.csv')
```

```python
# 상위 5개만 남기고 나머지는 '기타'로 처리하는 함수
def top_n_with_others(series, n=5):
    value_counts = series.value_counts()
    top_n = value_counts.head(n)
    others_sum = value_counts[n:].sum()
    
    if others_sum > 0:
        top_n['기타'] = others_sum
    
    return top_n

# 서브플롯 생성
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Brand 파이 차트
brand_data = top_n_with_others(df['Brand'], n=5)
axes[0].pie(brand_data.values, labels=brand_data.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Brand 분포 (상위 5개)', fontsize=14, fontweight='bold')

# Color 파이 차트
color_data = top_n_with_others(df['Color'], n=5)
axes[1].pie(color_data.values, labels=color_data.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Color 분포 (상위 5개)', fontsize=14, fontweight='bold')

# Bag_Type 파이 차트
bag_type_data = top_n_with_others(df['Bag_Type'], n=5)
axes[2].pie(bag_type_data.values, labels=bag_type_data.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title('Bag_Type 분포 (상위 5개)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
